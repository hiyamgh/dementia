```python
import pandas as pd
import shap
import sklearn
import numpy as np
from imblearn.ensemble import BalancedRandomForestClassifier
from collections import Counter
```

```python
# split into input & output
target_variable = 'dem1066'
cols_drop = None
df_train = pd.read_csv('../input/train.csv')
df_test = pd.read_csv('../input/test.csv')
    
# datasets of train/test split
X_train_d = df_train.loc[:, df_train.columns != target_variable]
y_train_d = df_train.loc[:, df_train.columns == target_variable]

X_test_d = df_test.loc[:, df_test.columns != target_variable]
y_test_d = df_test.loc[:, df_test.columns == target_variable]

# numpy nd-arrays of train/test split
X_train = np.array(df_train.loc[:, df_train.columns != target_variable])
y_train = np.array(df_train.loc[:, df_train.columns == target_variable])

X_test = np.array(df_test.loc[:, df_test.columns != target_variable])
y_test = np.array(df_test.loc[:, df_test.columns == target_variable])

counter = Counter(y_train_d[target_variable])
reverse_counter={0:counter[1],1:counter[0]}
reverse_counter
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
model = BalancedRandomForestClassifier()
model.fit(X_train, y_train)
explainer = shap.TreeExplainer(model)

# get the expected value 
expected_value = explainer.expected_value
print('expected value: {}'.format(expected_value))
if isinstance(expected_value, list):
    print(True)
    expected_value = expected_value[1]
print(f"Explainer expected value: {expected_value}")

shap_values = explainer.shap_values(X_test_d)
shap_interaction_values = explainer.shap_interaction_values(X_test)
print("shap values shape: {}".format(np.array(shap_values).shape))
print("expected values shape: {}".format(np.array(expected_value).shape))
```

```python
shap.summary_plot(shap_values[0], X_test_d, plot_type="violin")
```

```python
shap.summary_plot(shap_values[1], X_test_d, plot_type="violin")
```

```python
for name in X_test_d.columns:
    shap.dependence_plot(name, shap_values[1], X_test_d)
```

```python
# takes a couple minutes since SHAP interaction values take a factor of 2 * # features
# more time than SHAP values to compute, since this is just an example we only explain
# the first 2,000 people in order to run quicker
shap_interaction_values = shap.TreeExplainer(model).shap_interaction_values(X_test_d)
np.array(shap_interaction_values).shape
```

get the list of predictions (as far as I understood, the probabilitic predictions (probability of class))

```python
dementia_predictions = model.predict_proba(X_test)
```

Create a function that generates labels for the plot legend. _Tip: Include the predicted values in the legend labels to help distinguish the classes._

```python
def class_labels(row_index):
    return [f'Class {i + 1} ({dementia_predictions[row_index, i].round(2):.2f})' for i in range(2)]
```

```python
row_index = 2
shap.multioutput_decision_plot(list(expected_value), shap_values,
                               row_index=row_index, 
                               feature_names=list(X_test_d.columns), 
                               highlight=[np.argmax(dementia_predictions[row_index])],
                               legend_labels=class_labels(row_index),
                               legend_location='lower right')  
```

get the row indices of the miss-classified features

```python
y_pred = model.predict(X_test)
misclassified = [i for i in range(len(y_pred)) if y_pred[i] != y_test[i]]
misclassified
```

```python
shap.decision_plot(expected_value[1], shap_interaction_values[1][misclassified], X_test_d.loc[misclassified],
                   link='logit')
```
